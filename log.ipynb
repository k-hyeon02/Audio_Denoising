```python
import os
import pandas as pd
import matplotlib.pyplot as plt

LOG_PATH = "./saved/16/training_log.csv"
OUT_DIR = "./saved/16"

os.makedirs(OUT_DIR, exist_ok=True)

# 1. 로그 읽기 (| 구분)
df = pd.read_csv(LOG_PATH, sep="|")
df.columns = df.columns.str.strip()   # 공백 제거

# 2. epoch을 int로 변환
df["epoch"] = df["epoch"].astype(int)

# 3. Loss 그래프
plt.figure()
plt.plot(df["epoch"], df["train_loss"], label="Train Loss")
plt.plot(df["epoch"], df["val_loss"], label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training / Validation Loss")
plt.legend()
plt.grid(True)

loss_path = os.path.join(OUT_DIR, "loss_curve.png")
plt.savefig(loss_path)
plt.close()

# 4. PSNR 그래프
plt.figure()
plt.plot(df["epoch"], df["train_psnr"], label="Train PSNR")
plt.plot(df["epoch"], df["val_psnr"], label="Val PSNR")
plt.xlabel("Epoch")
plt.ylabel("PSNR (dB)")
plt.title("Training / Validation PSNR")
plt.legend()
plt.grid(True)

psnr_path = os.path.join(OUT_DIR, "psnr_curve.png")
plt.savefig(psnr_path)
plt.close()

print("Saved graphs:")
print(loss_path)
print(psnr_path)
```
