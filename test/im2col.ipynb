```python
import torch

H = 3
W = 3
padding = 1
stride = 2

kernel_h = 3
kernel_w = 3
```

```python
out_h = (H + 2 * padding - kernel_h) // stride + 1
out_w = (W + 2 * padding - kernel_w) // stride + 1
print(out_h, out_w)
```

```python
offset_h = torch.arange(kernel_h).reshape(-1, 1).repeat(1, kernel_w).reshape(-1)
print(offset_h)

start_h = torch.arange(out_h) * stride
start_h = start_h.reshape(-1, 1).repeat(1, out_w).reshape(-1)
print(start_h)
```

```python
offset_h = offset_h.reshape(-1,1)
print(offset_h)
```

```python
start_h = start_h.reshape(1, -1)
print(start_h)
```

```python
h_idx = offset_h + start_h
print(h_idx)
```
